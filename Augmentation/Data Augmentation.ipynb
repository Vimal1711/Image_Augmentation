```python
import tensorflow as tf
import cv2
import matplotlib.pyplot
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import keras.utils as image
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.1,
    shear_range=0.25,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
directory_path=r"C:\Users\Vimal\Desktop\ML\Agrobez\Dataset\Augmentation"
```

```python
# Generate augmented images
data = datagen.flow_from_directory(
    directory=directory_path,
    target_size=(224,224),
    color_mode="rgb",
    batch_size=32,
)
```

```python
print(data.directory)
```

```python
print(len(data))
```

```python
# Saving images
import os
for i in range(len(data)):
    images, labels = data[i]
    for j in range(len(images)):
        image = images[j]
        file_name = "aug4_" + str((i*10) * 32 + j) + ".jpg"
        image_path = os.path.join(directory_path, file_name)
        cv2.imwrite(image_path, image)
print("Completed Successfully!!")
```

```python
import matplotlib.pyplot as plt

output = ['original.jpg', 'rotated.jpg', 'horizontal_flip.jpg', 'vertical flip.jpg']
titles = ['Original Image', 'Rotated Image', 'Horizontal Flipped Image', 'Vertical flip Image']

for i in range(4):
    plt.subplot(2, 2, i + 1)  
    image = plt.imread(output[i])  
    plt.imshow(image)
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.tight_layout() 
```
